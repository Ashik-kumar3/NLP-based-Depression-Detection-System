```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import re
```

```python
import nltk
nltk.download('stopwords')
```

```python
stop_words = set(stopwords.words('english'))

from nltk.stem import WordNetLemmatizer

lem = WordNetLemmatizer()

def pre(text):
    
    stemmed = re.sub(r'[^a-zA-Z ]','',text)
    stemmed = stemmed.lower()
    stemmed = stemmed.split()
    stemmed = [lem.lemmatize(i,pos='v') for i in stemmed if i not in stopwords.words('english')]
    stemmed = ' '.join(stemmed)
    return stemmed
```

```python
df['message to examine'] = df['message to examine'].apply(pre)
df['message to examine']
```

```python
x = df['message to examine']
y = df['label (depression result)']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer()
vectorizer.fit(x)
x = vectorizer.transform(x)
```

```python
print(x)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.3,stratify=y,random_state=2)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(xtrain,ytrain)
ypred = model.predict(xtest)
ypred
```

```python
print(xtest)
```

```python
print(ypred)
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(ytest,ypred)
acc
```

```python
ytrain_pred = model.predict(xtrain)
```

```python
train_acc = accuracy_score(ytrain,ytrain_pred)
train_acc
```
